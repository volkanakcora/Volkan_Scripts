```python
import seaborn as sns
flights = sns.load_dataset('flights')
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df = df.pivot("month","year","passengers")
df.head()
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df, annot = True , fmt = 'd', linewidths = 0.5)
```
