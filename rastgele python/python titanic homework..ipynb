```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate

%matplotlib inline
```

```python
data = pd.read_csv("titanic3.csv")
```

```python
data.drop("cabin",inplace=True,axis=1)
```

```python
pd.get_dummies(data["sex"])
```

```python
sex = pd.get_dummies(data["sex"],drop_first=True)
```

```python
embarked = pd.get_dummies(data["embarked"],drop_first=True)
pclass = pd.get_dummies(data["pclass"],drop_first=True)
```

```python
def add_age(cols):
    age = cols[0]
    pclass = cols[1]
    if pd.isnull(age):
        return int(data[data["pclass"] == pclass]["age"].mean())
    else:
        return age
```

```python
data["age"] = data[["age", "pclass"]].apply(add_age,axis=1)
```

```python
data = pd.concat([data,pclass,sex,embarked],axis=1)
```

```python
data
```

```python
data.drop(["boat","body"],axis=1,inplace=True)
```

```python
def add_survived(cols):
    survived = cols[0]
   
    if pd.isnull(survived):
        return int(data[data == survived]["survived"].mean())
    else:
        return survived
```

```python
import numpy as np
data =data[~data.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
X = data.drop("survived",axis=1)
y = data["survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```
